```python
import numpy as np
import pandas as pd
import datetime as dt

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/drive/MyDrive/SeSAC.seoul/05_머신러닝 데이터분석/참고자료/traffic_clean_utf.csv")
df.head()
```

```python
feature = ['날짜', '기온', '강수량', '풍속', '습도', '일조', '적설', '지면온도']
label = ['사상자']
```

```python
df[feature].head()
```

```python
train_input = df[['기온', '강수량', '풍속', '습도', '일조', '적설', '지면온도']].to_numpy()
train_target = df['사상자'].to_numpy()
```

```python
train_input, test_input, train_target, test_target = train_test_split(train_input, train_target, test_size=0.2, train_size=0.8, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(train_scaled, train_target)
```

```python
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```
