### Step 1. 필요한 라이브러리 들을 import 하세요

```python
import pandas as pd
```

### Step 2. 다음 raw_data 를 이용하여 DataFrame을 만드세요

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'], 
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'], 
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'], 
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

### Step 3. regiment 라는 변수에 DataFrame을 할당 하세요.
#### 각 column값을 raw_data의 key값으로 설정하세요

```python
regiment = pd.DataFrame(raw_data, columns=raw_data.keys())
```

```python
regiment.head()
```

### Step 4. Nighthawks regiment의 평균 preTestScore는 얼마입니까?

```python
# 방법1. groupby
regiment.groupby("regiment").mean(numeric_only=True).loc["Nighthawks"].preTestScore
```

```python
# 방법2. filtering
regiment[(regiment.regiment == "Nighthawks")]["preTestScore"].mean()
```

### Step 5. company의 종합 통계량을 제시하세요

```python
regiment.groupby("company").describe()
```

### Step 6. 각 company의 평균 preTestScore는 얼마입니까?

```python
regiment.groupby("company").preTestScore.mean()
```

### Step 7. regiment 별 company의 평균 preTestScores 제시하세요

```python
regiment.groupby(["regiment","company"]).preTestScore.mean()
```

### Step 8. regiment 별 company의 평균 preTestScores를 heirarchical indexing 없이 제시하시오.
#### Hint : unstack() https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.unstack.html?highlight=unstack

```python
regiment.groupby(["regiment","company"]).preTestScore.mean().unstack(-2)
```

```python
regiment.groupby(["regiment","company"]).preTestScore.mean().unstack(-1)
```

### Step 9. regiment 별 company에 대해 다음과 같은 통계량을 제시하시오.
#### preTestScore, postTestScore의 평균

```python
regiment.groupby(["regiment","company"])[["preTestScore", "postTestScore"]].mean()
```

```python
regiment.groupby(["regiment","company"])[["preTestScore", "postTestScore"]].mean().unstack(-1)
```
