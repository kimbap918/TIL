### Step 4. "school"부터 "guardian"까지 DataFrame을 슬라이싱 하여 stud_alcoh에 할당하세요. 단 "guardian" column이 포함되어야 합니다. 

```python
stud_alcoh = df.loc[:, "school":"guardian"]
```

```python
stud_alcoh.head()
```

### Step 5. 문자열의 앞글자를 대문자로 변경해주는 lambda 함수 또는 함수를 만드세요

```python
"hello world".capitalize()
```

```python
"hello world".upper()
```

```python
capitalizer = lambda x:x.capitalize()
```

```python
capitalizer("hello world")
```

### Step 6. Mjob 와 Fjob column에 step5 에서 만든 function을 적용하고 원래 stud_alcoh DataFrame에 적용하세요

```python
stud_alcoh.Mjob = stud_alcoh.Mjob.apply(capitalizer)
stud_alcoh.Fjob = stud_alcoh.Fjob.apply(capitalizer)
stud_alcoh.head()
```

```python
stud_alcoh
```

### Step 7. legal_drinker 라는 Column을 만들고 다음과 같은 내용을 추가 합니다.
- 18세 이상은 True, 18세 미만은 False

```python
# stud_alcoh["legal_drinker"] = stud_alcoh.age >= 18
```

```python
def legal_drinker(x):
    if x>= 18:
        return True
    else:
        return False
stud_alcoh["legal_drinker"] = stud_alcoh.age.apply(legal_drinker)
```

```python
stud_alcoh.head()
```
